```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df1=pd.read_csv('test.csv')
```

```python
df.head()
```

```python
df.drop(['time','updated'],axis=1,inplace=True)
df1.drop(['time','updated'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['typeN']=lb.fit_transform(df['type'])
df['statusN']=lb.fit_transform(df['status'])
df['magtypeN']=lb.fit_transform(df['magType'])
df1['typeN']=lb.fit_transform(df1['type'])
df1['statusN']=lb.fit_transform(df1['status'])
df1['magtypeN']=lb.fit_transform(df1['magType'])
```

```python
df.drop(['type','status','magType'],axis=1,inplace=True)
df1.drop(['type','status','magType'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.fillna(df.median(),inplace=True)
df1.fillna(df.median(),inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df.drop(['place','id'],axis=1,inplace=True)
```

```python
model=DecisionTreeClassifier()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(['magSource'],axis=1),df.magSource,test_size=0.3)

```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
